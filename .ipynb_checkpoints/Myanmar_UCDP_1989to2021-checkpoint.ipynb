### Bokeh

```python
# reset index so "date_end" can be used for bokeh
UCDP_Myanmar = UCDP_Myanmar.reset_index()
```

```python
output_notebook()
```

The graph below shows the number of deaths from all the events verified by UCDP, each point is an event, and days where events have not been recorded do not have points that show up on this graph.

To make the linegraph more accurate, the data needs to be cleaned further. Dates which haven't been included (as there is no conflict event recorded by UCDP), will need to be included as 0 or NaN. 

From comparing the length of 'idx' below and the shape of UCDP_Myanmar, we can clearly see that since UCDP records began, there has been a recorded conflict event for more than 1/3 of the total days. 

```python
# period_range: Return a fixed frequency PeriodIndex. The day (calendar) is the default frequency.
# this returns each day between the period of the 1st recorded date and the last recorded date.
idx = pd.period_range(min(UCDP_Myanmar['date_end']), max(UCDP_Myanmar['date_end']))

print(idx)

UCDP_Myanmar.shape
```

```python
idxV2 = pd.date_range(min(UCDP_Myanmar['date_end']), max(UCDP_Myanmar['date_end']))

print(idxV2)
```

```python
# we now need to check idxV2 to see if there are missing dates in UCDP_Myanmar["date_end"], then append new 0 filled rows for 
# each date missing from UCDP_Myanmar, the UCDP_Myanmar["date_end"] column will be filled with whatever date was found to be missing

# adds new row with 0 values
# UCDP_Myanmar.loc[len(df)] = 0
# change specific cell value to the missing value from idxV2

a = 0
for x in idxV2:
    if x in UCDP_Myanmar["date_end"]:
        a += 1

print(a)

UCDP_Myanmar["date_end"]
```

```python
# create a new DataFrame containing all the expected dates
date_range = pd.date_range(min(UCDP_Myanmar['date_end']), max(UCDP_Myanmar['date_end']))
expected_df = pd.DataFrame({'date_end': date_range})

# join the two dataframes on the 'date_end' column using a left join
merged_df = pd.merge(expected_df, UCDP_Myanmar, on='date_end', how='left')

# replace missing values with 0
merged_df = merged_df.fillna(0)

# print the resulting dataframe
merged_df
```

```python
merged_df.dtypes
```

```python
p2 = figure(x_axis_type="datetime", title="Recorded Deaths per Event", width=800)
p2.grid.grid_line_alpha=0.3
p2.xaxis.axis_label = "Conflict Event by Date"
p2.yaxis.axis_label = "Deaths"

p2.line(merged_df["date_end"], merged_df["best"], color='#f2a900', legend='Number of Deaths')
#p2.line(eth.index, eth['ClosePrice'], color='#A6CEE3', legend='Ether')

p2.legend.location = "top_left"

show(p2)
```

### Aggregate Data
Aggregate the death figures by month.

```python
#merged_df["date_end"].dt.strftime('%Y-%m') extracts the year and month from the dates in column "date_end" and returns a string
#in the format 'YYYY-MM'.
#merged_df.groupby() groups the rows in the dataframe by the strings obtained from column "date_end".
#["deaths_a", "deaths_b", "deaths_civilians", "deaths_unknown", "best", "high", "low"].sum() applies the sum() method to several
#columns for each group.
#The output will be a Pandas Dataframe with the sums of values in columns "deaths_a", "deaths_b", "deaths_civilians", 
#"deaths_unknown", "best", "high", "low" for each month.
#If we were only using 2 columns, e.g. "date_end" and "best", the result would be a Series, and we would need to add
#.to_frame() at the end to convert it to a dataframe.

UCDP_monthly = merged_df.groupby(merged_df["date_end"].dt.strftime("%Y-%m"))["deaths_a", "deaths_b", "deaths_civilians", 
                                                                             "deaths_unknown", "best", "high", "low"].sum()

UCDP_monthly = UCDP_monthly.reset_index()

#here we can see "date_end" has been converted to a string
UCDP_monthly.dtypes
```

```python
#bokeh will need the x axis to be numerical or something with logical order, so we will convert "date_end" back to datetime.
UCDP_monthly["date_end"]= pd.to_datetime(UCDP_monthly["date_end"])

UCDP_monthly
```

```python
p3 = figure(x_axis_type="datetime", title="Recorded Deaths per Month", width=800)
p3.grid.grid_line_alpha=0.3
p3.xaxis.axis_label = "Conflict Event by Month"
p3.yaxis.axis_label = "Deaths"

p3.line(UCDP_monthly["date_end"], UCDP_monthly["best"], color='#f2a900', legend='Total Number of Deaths')

p3.legend.location = "top_left"

show(p3)
```
